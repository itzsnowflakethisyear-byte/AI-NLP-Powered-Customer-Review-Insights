```python
reviews_df = reviews_df[
    reviews_df["categories"].str.contains(r"\bRestaurants\b", na=False)
]
```

```python
def extract_cuisine(cat_str):
    if not isinstance(cat_str, str) or not cat_str:
        return None
    # Split categories and strip whitespace
    cats = [c.strip() for c in cat_str.split(",")]
    # Remove 'Restaurants' and return the first remaining category, if any
    for c in cats:
        if c.lower() != "restaurants":
            return c
    return None  # If only 'Restaurants' was present

# Apply to the dataframe
reviews_df["cuisine"] = reviews_df["categories"].apply(extract_cuisine)
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

model = SentenceTransformer('all-MiniLM-L6-v2')

# Known cuisine seeds (expandable)
seed_cuisines = [
    'Italian', 'Chinese', 'Mexican', 'Indian', 'Thai', 'Japanese', 'French',
    'Mediterranean', 'American', 'Korean', 'Vietnamese', 'Greek', 'Spanish',
    'Lebanese', 'Persian', 'Brazilian', 'Peruvian', 'Senegalese', 'Poutine'
]

seed_embeddings = model.encode(seed_cuisines)
test_embeddings = model.encode(reviews_df['cuisine'].astype(str).tolist())

similarities = cosine_similarity(test_embeddings, seed_embeddings)
reviews_df['food_similarity'] = np.max(similarities, axis=1)

# Threshold based on validation (typically 0.35-0.45 works)
reviews_df['is_food'] = reviews_df['food_similarity'] > 0.4
```
